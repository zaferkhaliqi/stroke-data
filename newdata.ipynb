```python
import matplotlib.pyplot as plt
```

```python
data2 =pd.read_csv("healthcare-dataset-stroke-data.csv")
data2
```

```python
iD  = data2["id"]
```

```python
data2  = data2.drop("id",axis = 1)#id ise yaramadigi icin atiyorum
```

```python
dummies = pd.get_dummies(data2[["gender","smoking_status","work_type"]],prefix  = ("gender","smoking_status","work_type")) #
data2 = pd.concat([data2,dummies],axis = 1)
data2
```

```python
del data2["gender"]
del data2["smoking_status"]
del data2["work_type"]
```

```python
from sklearn.preprocessing import LabelEncoder
LB = LabelEncoder()

data2["ever_married"] = LB.fit_transform(data2["ever_married"])#marride = 1 else = 0
data2["Residence_type"] =  LB.fit_transform(data2["Residence_type"])#Rural = 0 urban = 1
```

```python
from sklearn.impute import KNNImputer
imputer  = KNNImputer() # imputer fonksyonu kayip verilerin yerine diger verilerin ortalamasina bakarak bi deger olusuturuyor
imputer.fit(data2)

data2 =pd.DataFrame(imputer.transform(data2),index = data2.index,columns = data2.columns)
data2
```

```python
import seaborn as sns
plt.figure(figsize = (40,20))
sns.heatmap(data2.corr(),annot=True,cmap = "RdYlGn")#burda korelasyon yaptim
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = data2[['age','hypertension','heart_disease','ever_married','avg_glucose_level']]

y = data2['stroke']
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.3,random_state=1)
scaler.fit(xTrain)
scaler.fit(xTest)
#datanin tumu yerine sadece x lerin scaler yapmamin nedeni = stroke degerlerinin degismemesi
xTrain =pd.DataFrame(scaler.transform(xTrain),index = xTrain.index,columns = xTrain.columns)
xTest =pd.DataFrame(scaler.transform(xTest),index = xTest.index,columns = xTest.columns)
xTrain
```

```python
x['stroke']= y
plt.figure(figsize = (40,20))
sns.pairplot(x,hue= 'stroke')#burda genel grafik analizi yaptim
```

Train test

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

```

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
logR.fit(xTrain,yTrain)
logRpredict = logR.predict(xTest)
logRaccuracy_scoree = accuracy_score(yTest, logRpredict)
logRCMscore = confusion_matrix(yTest, logRpredict)#confusion_matrix#DecisionTreeClassifier
print("confusion_matrix score LogisticRegression = ")
print(logRCMscore)
sns.heatmap(logRCMscore, annot=True)

print("accuracy_score score LogisticRegression = ",)
print(logRaccuracy_scoree)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()#DecisionTreeClassifier
DTC.fit(xTrain,yTrain)
DTCpredict = DTC.predict(xTest)

DTFCMscore = confusion_matrix(yTest, DTCpredict)#confusion_matrix#DecisionTreeClassifier
DTFaccuracy_scoree = accuracy_score(yTest, DTCpredict)
print(DTFCMscore)

sns.heatmap(DTFCMscore, annot=True)

print("accuracy_score score DecisionTreeClassifier = ")
print(DTFaccuracy_scoree)
```
